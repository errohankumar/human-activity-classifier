```python
import os
import warnings
import sys 

from glob import glob

import re
import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
import sklearn

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 


import pickle
import errno
```

```python
## Let us read from the pickled file 
pickle_in = r'/Users/rohankumar/Desktop/AutoScout24/Case Study/Data/pickle/file_1.pickle'
pl = open(pickle_in, "rb")
df = pickle.load(pl)
df
```

### Distribution of Make Name 

```python
# set the figure size and position
fig, ax = plt.subplots(figsize=(15,20))

#title
plt.title("Distribution of Brands according to their total counts", fontsize=20)

ax = sns.barplot(y = df.make_name.unique(), x = df.make_name.value_counts(), data=df)
ax.tick_params(labelrotation=0, labelsize=12)
#df['make_name'].plot()
```

### Distribution of product_tier

```python
# set the figure size and position
fig, ax = plt.subplots(figsize=(14,8))

#title
plt.title("Distribution of Product tier according to their total counts", fontsize=20)

ax = sns.barplot(x = df.product_tier.unique(), y = df.product_tier.value_counts(), data=df)
ax.tick_params(labelrotation=0, labelsize=12)
#df['make_name'].plot()
```

<p> The barplot for the Product Tier and total counts shows that the dataset is pretty unbalanced. There are more brands in the basic tier than in the premium and plus. This can make prediction of <b>Product Tier</b> difficult! </p>

### Distribution of make_name with price

```python
# set the figure size and position
fig, ax = plt.subplots(figsize=(14,30))

#title
plt.title("Distribution of make_name with price", fontsize=20)

ax = sns.barplot(y = df.make_name, x = df.price, data=df)
ax.tick_params(labelrotation=0, labelsize=14)
ax.secondary_xaxis('top')
```

### Distribution of price with product_tier

```python
# set the figure size and position
fig, ax = plt.subplots(figsize=(14,7))

#title
plt.title("Distribution of price with product_tier", fontsize=20)

ax = sns.boxplot(y = df.price, x = df.product_tier, data=df)
#ax.tick_params(labelrotation=0, labelsize=14)
```

### Distribution of search views and detail views with 'first_registration_year'

```python
# set the figure size and position
fig, ax = plt.subplots(1,2, figsize=(14,7))

#title
plt.suptitle("Distribution of Views with Product-Tier", fontsize=20)

# Search-Views
sns.lineplot(y = df.search_views, x = df.first_registration_year, data=df, ax=ax[0])

# Detail-Views
sns.lineplot(y = df.detail_views, x = df.first_registration_year, data=df, ax=ax[1])
```

```python
fig, ax = plt.subplots(1,2, figsize=(14,7))

#title
plt.suptitle("Distribution of Price with Registration Year", fontsize=20)

sns.lineplot(y = df.price, x = df.first_registration_year, data=df, ax=ax[0])

# Detail-Views
sns.lineplot(y = df.stock_days, x = df.first_registration_year, data=df, ax=ax[1])
```

```python
fig, ax = plt.subplots(figsize=(14,7))

#title
plt.suptitle("Distribution of Price with Registration Year", fontsize=20)

sns.boxplot(y = df.price, x = df.product_tier, data=df)
```
