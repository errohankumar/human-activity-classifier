<h3>Feature Engineering</h3>
<p>The feature <b>first_zip_digit</b> is not useful and we can safely remove it. Also, the feature <b>stock_days</b> has negative values. This is due to incorrect calculation of the stock days from <b>created_date</b> and <b>deleted_date</b>. We have to re-calculate the stock_days feature and add it back to our dataset.</p>

```python
df.head()
```

```python
## Re-calculate the stock_days column 
df['stock_days'] = df['deleted_date'] - df['created_date']
df['stock_days'] = df['stock_days'].dt.days.astype(int)

## Drop the first_zip_digit feature
df = df.drop(columns={'first_zip_digit'})

## Reset the index
df = df.reset_index(drop=True)

## A new feature can be introduced called age of the vehicle or time_since_registration
'''
The age of the vehicle in days is basically the created_date - first_registration_year. This value will provide further 
insight on how the price is affected. I am taking an assumption that the price of the vehicle should be less for 
the older vehicles. 
'''
df['time_since_registration'] = df['created_date'] - df['first_registration_year']
df['time_since_registration'] = df['time_since_registration'].dt.days.astype(int)
df.describe()
```

### Important !
<p>The <b>time_since_registration</b> has negative values. This is because there are rows where the created_date
is before the <b>first_registration_year</b>. When I looked into it, this exists for all the make-names in the year
2019. The created_date is few months before the registration year, which likely causes doubt since it might be an
error or could be because of some policy. This question needs to be answered!</p>

```python
# set the figure size and position
fig, ax = plt.subplots(figsize=(20,12))

#title
plt.title("Distribution of make_name vs time_since_registration(Age of the car in Days)", fontsize=20)

ax = sns.barplot(x = df.make_name, y = df.time_since_registration, data=df)
ax.tick_params(labelrotation=90, labelsize=12)
plt.xlabel('Make Name', size=15)
plt.ylabel('Age (Days)', size=15)
```
