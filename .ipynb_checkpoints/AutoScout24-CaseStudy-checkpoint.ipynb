```python
# Let us match all the object elements and see if there are other values than just numeric
i = 0
list_numbers = []
while(i<df_case_study_new.shape[0]):
    if(bool(re.search(r"^\d*\.?\d*$", str(df_case_study_new['ctr'].loc[i])))== False):
        list_numbers.append(df_case_study_new['ctr'].loc[i])
        print(df_case_study_new['ctr'].loc[i])
    i += 1
```

<p>We can see from the above list that the ctr has some strange values! 01 April which is weirdly present in the dataset!</p>

```python
## Let us create the dataset of values in the ctr that are absurd
absurd_dataset=pd.DataFrame()
for element in list_numbers:
    absurd_dataset = absurd_dataset.append(df_case_study_new[df_case_study_new['ctr'] == element])
absurd_dataset
```

<p> The rows with absurd or strange values in the <b>ctr</b> needs to be dropped since they vary  significantly from the other values in the dataset. These values are significantly higher and I am assuming here that they are inherently incorrect.</p>

```python
## Dropping the rows using the index in the 'absurd dataset'

df_case_study_new = df_case_study_new.drop(index=absurd_dataset.index)

# Reset the index 
df_case_study_new = df_case_study_new.reset_index(drop=True)
df_case_study_new
```

### 4. Set appropriate datatypes for each column 

This step would allow us to carry on with the further analysis. The datatypes will be set according to column types.

```python
# Let us drop the article_id column since it has no importance 
df_case_study_new = df_case_study_new.drop(columns={'article_id'})
df_case_study_new
```

```python
df_case_study_new.dtypes
```

```python
## Look at the first 5 rows and change the datatype accordingly
import re

def set_datatype(df):
    # select the first 5 rows as samples
    sample_df = df.iloc[:5,:]
        
    i = 0 
    # Run until total number of columns are reached
    while(i< sample_df.shape[1]):
        c,d,e = 0,0,0
        sample_list = sample_df.iloc[:,i].tolist()
        print(sample_list)
        for element in sample_list:
            if(bool(re.search("^[a-zA-Z-\s]*$", str(element))) == True):
                # String counter
                c += 1
            elif(bool(re.search(r"^\d*\.?\d*$", str(element)))== True):
                # Numeric counter
                d += 1
            elif (bool(re.search(r"[0-3]?[0-9].[0-3]?[0-9].(?:[0-9]{2})?[0-9]{2}$", str(element))) == True):
                # Date counter
                e += 1
            else:
                print('Unknown DataType')
            
        # check if c and d equals the length of the dataset
        df = change_datatype(c, d, e, i, df, sample_df) 
        i += 1
    return df
 
# Change the datatype based on the key 
def change_datatype(c, d, e, i, df, sample_df):
    if c == sample_df.shape[0]:
        # Change the datatype to string
        df.iloc[:,i] = df.iloc[:,i].astype(str)
    elif d == sample_df.shape[0]:
        print('inside d')
        # Change the datatype to numeric
        df.iloc[:,i] = df.iloc[:,i].astype(float)
        
    elif e == sample_df.shape[0]:
        # Change the datatype to datetime
        df.iloc[:,i] = pd.to_datetime(df.iloc[:,i], format="%d.%m.%y")
    else:
        print('Datatypes cannot be set')
    return df

if __name__ == '__main__':
    final_df = set_datatype(df_case_study_new)
    print("Successfully set the datatypes")

## The 'first_registration_year' column is now a float datatype. Let's put it to datetime type!

final_df['first_registration_year'] = pd.to_datetime(final_df['first_registration_year'], format = '%Y')
final_df['first_registration_year']
```

#### 5.1 Exploring the price feature

```python
price_range_0_50 = [0, 50]
price_range_50_100 = [50, 100]
price_range_100_200 = [100, 200]
price_range_200_300 = [200, 300]
price_range_300_400 = [300, 400]
price_range_400_500 = [400, 500]
price_range_500_600 = [500, 600]

price_range_100000_200000 = [100000, 200000]
price_range_200000_300000 = [200000, 300000]


price_ranges = [price_range_0_50, price_range_50_100, price_range_100_200, price_range_200_300,
                price_range_300_400, price_range_400_500, price_range_500_600,
               price_range_100000_200000, price_range_200000_300000]

for price_range in price_ranges:
    print('Total number of vehicles sold between {}â‚¬ and {}â‚¬ are {}'.format(price_range[0],price_range[1],
        (final_df[(final_df.price >= price_range[0]) & (final_df.price<= price_range[1])].price.count())))
```

```python
## Distribution of price vs make-name

fig, ax = plt.subplots(figsize=(20,10))
ax = sns.boxplot(y=final_df['price'], x=final_df['make_name'], data=final_df)
ax.tick_params(labelrotation=90, labelsize=14)
plt.title('Distribution of Price vs Make Name')
plt.ylabel('Price', size=14)
plt.xlabel('Make Name', size=14)
```

<p>From the price vs make-name graph, we can see many outliers. It will be beneficial if we get rid of these outliers in the dataset. To do this, there are two methods namely:</p>

<ol>
    <li>Z-Score</li>
    <li>IQR</li>
</ol>

<p>I will be using IQR since it is more robust for outliers. I have tried using Z-score but did not find the results promising.</p>

```python
## Remove using IQR
Q1 = final_df['price'].quantile(0.25)
Q3 = final_df['price'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (final_df['price'] >= Q1 - 1.5 * IQR) & (final_df['price'] <= Q3 + 1.5 *IQR)
final_df_copy = final_df.loc[filter] 
final_df_copy.describe()
```

```python
## After Outlier removal in Price feature

## Plot of Price vs Make-Name
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.boxplot(y=final_df_copy['price'], x=final_df_copy['make_name'], data=final_df_copy)
ax.tick_params(labelrotation=90, labelsize=14)
plt.title('Distribution of Price vs Make Name after outlier removal')
plt.ylabel('Price', size=14)
plt.xlabel('Make Name', size=14)
```

```python
## Scatterplot for price vs Make-Name
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.scatterplot(y=final_df_copy['price'], x=final_df_copy['make_name'], data=final_df_copy)
ax.tick_params(labelrotation=90, labelsize=14)
plt.title('Distribution of Price vs Make Name after outlier removal')
plt.ylabel('Price', size=14)
plt.xlabel('Make Name', size=14)
```
