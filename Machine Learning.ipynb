### Grid Search and Hyperparamer Tuning

This can be implemented and paramters cen be optimally selected

```python
pipeline_lr = Pipeline([('scaler_1',StandardScaler()),('1_lr',LinearRegression())])

pipeline_svr = Pipeline([('scaler_2',StandardScaler()),('kernel',Nystroem(random_state=1, n_components=100)),
                         ('1_svr',LinearSVR())])
                         
pipeline_rf = Pipeline([('1_rf',RandomForestRegressor())])
                         

model_pipeline = [pipeline_lr, pipeline_svr, pipeline_rf]
                         
# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0:'Linear Regression', 1: 'SVR', 2: 'RandomForest'}
                         
parameters = [
    {},
    {
    '1_svr__epsilon': [1e-3, 1e-4],
    '1_svr__C':[1, 10.0, 100.0, 1000.0]
    },
    {
    '1_rf__n_estimators': [75],
    '1_rf__max_depth' : [2,4,6],
    '1_rf__criterion' : ["mae","mse"]
    }
]
                         
# fit the pipeline with the training data
for model, pipe in zip(model_pipeline, pipe_dict):
    grid_search = GridSearchCV(estimator=model, 
                               param_grid=parameters[pipe], 
                               cv = StratifiedKFold(n_splits=10,shuffle=True,random_state=42),
                               scoring='neg_mean_absolute_error')
    grid_search.fit(train_features, train_labels)

    joblib.dump(grid_search, f"/Users/rohankumar/Desktop/AutoScout24/Case Study/Data/trained_models/{pipe_dict[pipe]}.pkl")

    print("{}: model best parameters are --> {}".format(pipe_dict[pipe], grid_search.best_params_))

    print("{}: cv accuracy is  {}".format(pipe_dict[pipe], grid_search.best_score_))
```
